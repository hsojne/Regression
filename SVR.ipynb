# SVR
- Support Vector Regression

> min $_{w, b, \zeta, \zeta^{*}} \frac{1}{2} ||w||^{2} + C \sum^{n}_{i=1} {( \zeta_{i} + \zeta_{i}^{*}}) $

> subject $ y_{i} - (w^{T} \phi(x_{i}) + b) \leq \epsilon + \zeta_{i} $ and
$ (w^{T} \phi(x_{i}) + b) - y_{i} \leq \epsilon + \zeta^{*}_{i} $ and
$ \zeta_{i} \zeta_{i}^{*} \geq 0 for 1 \leq i \leq n $

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# data
w0 = 2.3
w1 = 3.5
num_data = 100
noise = np.random.normal(0, 6 , num_data)

x = np.linspace(0, 10, num_data)
y = w0 + w1 * x + noise

# plot
plt.scatter(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.margins(0.2)
```

```python
# SVR
from sklearn.svm import SVR

epsilon = 5

svr_linear = SVR(kernel="linear", epsilon=epsilon)
svr_poly = SVR(kernel="poly", degree=3, epsilon=epsilon)
svr_rfb = SVR(kernel="rbf", epsilon=epsilon)
svr_sigmoid = SVR(kernel="sigmoid", epsilon=epsilon)

svr_linear.fit(x.reshape(-1, 1), y)
svr_poly.fit(x.reshape(-1, 1), y)
svr_rfb.fit(x.reshape(-1, 1), y)
svr_sigmoid.fit(x.reshape(-1, 1), y)


# plot process
x_range = np.linspace(x.min(), x.max(), 100).reshape(-1, 1)

y_linear = svr_linear.predict(x_range)
y_poly = svr_poly.predict(x_range)
y_rfb = svr_rfb.predict(x_range)
y_sigmoid = svr_sigmoid.predict(x_range)

plt.figure(figsize=(12, 12))


# Linear kernel
plt.subplot(2, 2, 1)
plt.scatter(x, y)
plt.plot(x_range, y_linear, color="red")
plt.fill_between(x_range.ravel(), y_linear - svr_linear.epsilon,
                 y_linear + svr_linear.epsilon,
                 color="red", alpha=0.2)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear")


# Polynomial kernel
plt.subplot(2, 2, 2)
plt.scatter(x, y)
plt.plot(x_range, y_poly, color="red")
plt.fill_between(x_range.ravel(), y_poly - svr_poly.epsilon,
                 y_poly + svr_poly.epsilon,
                 color="red",  alpha=0.2)
plt.xlabel("x")
plt.ylabel("y")
plt.title("poly")

# RFB kernel
plt.subplot(2, 2, 3)
plt.scatter(x, y)
plt.plot(x_range, y_rfb, color="red")
plt.fill_between(x_range.ravel(), y_rfb - svr_rfb.epsilon,
                 y_rfb + svr_rfb.epsilon,
                 color="red",  alpha=0.2)
plt.xlabel("x")
plt.ylabel("y")
plt.title("RFB")

# Sigmoid kernel
plt.subplot(2, 2, 4)
plt.scatter(x, y)
plt.plot(x_range, y_sigmoid, color="red")
plt.fill_between(x_range.ravel(), y_sigmoid - svr_sigmoid.epsilon,
                 y_sigmoid + svr_sigmoid.epsilon,
                 color="red",  alpha=0.2)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Sigmoid")











```
